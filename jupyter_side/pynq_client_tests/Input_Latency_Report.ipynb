# Button RTT Report

Notebook-first workflow for capturing button-triggered RTT samples from the PYNQ board and plotting them for the report.

This notebook uses the same GPIO button path as `run_pynq.py` and records one UDP RTT sample for each new button press.

```python
from pathlib import Path
import subprocess
import sys

NOTEBOOK_DIR = Path.cwd()
if (NOTEBOOK_DIR / "run_pynq_test.py").exists():
    PACKAGE_PARENT = NOTEBOOK_DIR.parent
else:
    PACKAGE_PARENT = NOTEBOOK_DIR
if str(PACKAGE_PARENT) not in sys.path:
    sys.path.insert(0, str(PACKAGE_PARENT))

from pynq_client_tests.plot_udp_rtt_csv import load_rtt_rows, plot_rtt_rows, summarise_by_label
```

## Configure Capture

Set the server, load label, and how many button presses you want to capture.

```python
SERVER = "3.9.71.204"
PORT = 9000
LABEL = "button-rtt"
SAMPLES = 20
CSV_PATH = Path(f"button_rtt_{LABEL}.csv")
JSON_PATH = Path(f"button_rtt_{LABEL}.json")

SERVER, PORT, LABEL, SAMPLES, CSV_PATH, JSON_PATH
```

## Capture CSV + JSON

Run this cell, then press PYNQ board buttons. Each new button press records one RTT sample.

```python
SCRIPT_DIR = NOTEBOOK_DIR if (NOTEBOOK_DIR / "run_pynq_test.py").exists() else NOTEBOOK_DIR / "pynq_client_tests"

cmd = [
    sys.executable,
    str(SCRIPT_DIR / "run_pynq_test.py"),
    "--server", SERVER,
    "--port", str(PORT),
    "--label", LABEL,
    "--samples", str(SAMPLES),
    "--csv-out", str(CSV_PATH),
    "--json-out", str(JSON_PATH),
]

print("Running:", " ".join(cmd))
subprocess.run(cmd, check=True)
```

## Compare Multiple Runs

Add more CSV files here if you want to compare different load labels.

```python
csv_files = [
    str(CSV_PATH),
    # "button_rtt_idle.csv",
    # "button_rtt_dual-board.csv",
]

rows = load_rtt_rows(csv_files)
len(rows)
```

## Summary Table

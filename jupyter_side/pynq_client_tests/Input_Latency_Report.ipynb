# Input Latency Report

Notebook-first workflow for capturing board input latency and plotting it for the report.

This notebook collects CSV samples from the monitor `/api/state` endpoint and then plots:
- average input -> server latency
- average input -> broadcast latency
- worst-case input -> broadcast latency
- per-sequence input -> broadcast trace


```python
from pathlib import Path
import subprocess
import sys

from pynq_client_tests.plot_input_latency_csv import load_latency_rows, plot_latency_rows, summarise_by_label
```

## Configure Capture

Set the monitor URL, load label, and how many distinct button-input samples you want to capture.

```python
MONITOR_URL = "http://3.9.71.204:8080"
LABEL = "idle"
SAMPLES = 40
CSV_PATH = Path(f"input_latency_{LABEL}.csv")

MONITOR_URL, LABEL, SAMPLES, CSV_PATH
```

## Capture CSV

Run this cell, then press buttons on the PYNQ board while the game is live.

```python
cmd = [
    sys.executable,
    "pynq_client_tests/input_latency_csv.py",
    "--monitor", MONITOR_URL,
    "--label", LABEL,
    "--samples", str(SAMPLES),
    "--csv-out", str(CSV_PATH),
]

print("Running:", " ".join(cmd))
subprocess.run(cmd, check=True)
```

## Load One Or More CSV Runs

Add extra CSV files here if you want to compare loads like `idle`, `monitor-open`, or `dual-board`.

```python
csv_files = [
    str(CSV_PATH),
    # "input_latency_monitor-open.csv",
    # "input_latency_dual-board.csv",
]

rows = load_latency_rows(csv_files)
len(rows)
```

## Summary Table

```python
summary = summarise_by_label(rows)
summary
```

## Plots

These render inline in Jupyter if `matplotlib` is available.

```python
figures = plot_latency_rows(rows)
figures
```
